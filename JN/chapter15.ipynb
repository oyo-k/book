```python
import numpy as np
win_count =  0 # 
while win_count <= 2 : 
    my_hand = input('my_hand: rock, scissors, or, paper : ')
    cpu_select = np.random.randint(3)
    if cpu_select == 0:
        cpu_hand = 'paper'
    elif cpu_select == 1:
        cpu_hand = 'scissors'
    else:
        cpu_hand = 'rock'    
    print('my_hand is ', my_hand, '. cpu_hand is ', cpu_hand)
    if my_hand == cpu_hand:
        print('***rock, paper, scissors***')
    elif my_hand == 'rock' and cpu_hand == 'scissors':
        win_count +=1 # 
        print('***You win. win_count', win_count ,'***')
    elif my_hand == 'scissors' and cpu_hand == 'paper':
        win_count +=1 # 
        print('***You win. win_count', win_count ,'***')
    elif my_hand == 'paper' and cpu_hand == 'rock':
        win_count +=1 # 
        print('***You win. win_count', win_count ,'***')        
    else:
        print('***You lose.***')
print('*******Game Over*******')
```
