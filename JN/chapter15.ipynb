```python
import pandas as pd
# データの読み込み
data = pd.read_csv('data//ken_data.csv', 
    encoding='shift_jis', index_col='郵便番号')
number = input('7桁郵便番号(-は不要)を入力してください:')
print('あなたの入力した郵便番号：',number, 'です。')
# whileによる繰り返し
i = 0
while data.index.values[i] != int(number):
    print('現在',i,'番目を検索中です')
    i +=1 # データフレームのインデックスを変化させるため
# 繰り返しを抜けて、出力
print('あなたの入力した郵便番号は',i,'番目にありました。') # 640820
print('郵便番号',number,'は',data.iloc[i,:],'です。')
```

```python
flag = True # 正しい入力まではTrueをFalseに変更しない

while flag: # flagがTrueなら不正な入力のため繰り返す
  # 正しい入力例は「640820」や「9071801」 
  number = input('7桁郵便番号(-不要)を入力してください:')
  try: # エラーが起きるまで下記のブロックを実行
      print('あなたの入力した郵便番号：',int(number), 'です。')
      flag = False # 繰り返しを抜けるためFalseに変更
  except ValueError: # エラーへの対処
      # エラーが起きた時点で、ここのブロックを実行
      print('文字ではなく数字を入力してください')

i = 0 # インデックスを参照するために必要な変数
while data.index.values[i] != int(number):
    print('現在',i,'番目を検索中です')
    i +=1 # 次のインデックスを参照するために加算    
# 繰り返し処理が終われば目的の住所を発見    
print('あなたの入力した郵便番号は',i,'番目にありました。') 
print('郵便番号',number,'は',data.iloc[i,:],'です。')
```

```python
import numpy as np
win_count =  0 # 読者が勝利した際のカウント
while win_count <= 2 : 
    my_hand = input('my_handの入力: rock, scissors, or, paper : ')
    cpu_select = np.random.randint(3)
    if cpu_select == 0:
        cpu_hand = 'paper'
    elif cpu_select == 1:
        cpu_hand = 'scissors'
    else:
        cpu_hand = 'rock'    
    print('my_hand is ', my_hand, '. cpu_hand is ', cpu_hand)
    if my_hand == cpu_hand:
        print('***rock, paper, scissors***')
    elif my_hand == 'rock' and cpu_hand == 'scissors':
        win_count +=1 # コンピュータに勝利した回数を加算
        print('***You win. win_count', win_count ,'***')
    elif my_hand == 'scissors' and cpu_hand == 'paper':
        win_count +=1 # コンピュータに勝利した回数を加算
        print('***You win. win_count', win_count ,'***')
    elif my_hand == 'paper' and cpu_hand == 'rock':
        win_count +=1 # コンピュータに勝利した回数を加算
        print('***You win. win_count', win_count ,'***')        
    else:
        print('***You lose.***')
print('*******Game Over*******')
```

```python
from turtle import *
penup() # ゴールライン描画開始 
speed(1) # ゆっくり進む
pencolor('red')
goto(200,-50) # ゴールライン
pendown()
left(90) # 上に向き
forward(100) #(200,50)まで線を引く
# A. ゴールライン描画終了
penup()    
home() # 亀を初期値(0, 0)に戻す 
pencolor('blue')
shape('turtle') #実行結果B. まで終了
print('現在亀は', position(), 'です')
x = xcor() # 判定用の亀のx座標
y = ycor() # 判定用の亀のy座標
# not(True) は False となる
while not(200 < x and -50 < y < 50):
  #実行結果C.のように
  forward(10) #線を越えるまで亀が進む
  dot()       #点をプロット
  x = xcor()  #移動したので位置情報を更新
  y = ycor()  #移動したので位置情報を更新
  print('亀の現在地', position())
done() #実行結果D.のように線を超える 
```
