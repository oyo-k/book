```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as ani

# figsizeとnの数値を大きくすればより鮮明な動画になる
fig = plt.figure(figsize=(2,2),dpi=150) # 画像の設定
n = 50        # 画像(配列)の大きさ
mod_para = 30 # モデルのパラメータ(配列の要素は0から29)
ims = []      # 描画した画像を一枚一枚保存する入れ物
old_w = np.zeros([n, n]) # size行size列の全て0の行列
# 行番号と列番号の中央の値のみが1の二次元配列
old_w[int(n/2),int(n/2)] = 1 # 初期状態

for step in np.arange(100): # 遷移の回数分繰り返し
  new_w = np.zeros([n, n])  # 次の状態の入れ物を用意
  for i in np.arange(n):    # 行の移動
    for j in np.arange(n):  # 列の移動
      # 現在の位置i と j の上下左右の値を加算する
      new_w[i,j] = old_w[i, j]            # 中央
      new_w[i,j] += old_w[i - 1, j]       # 上の状態
      new_w[i,j] += old_w[(i + 1) % n, j] # 下の状態
      new_w[i,j] += old_w[i, j - 1]       # 左の状態
      new_w[i,j] += old_w[i, (j + 1) % n] # 右の状態 
      new_w[i,j] %= mod_para # 合計で割ったあまりを計算
      
  # 遷移した状態の各要素の値を使い一枚の画像に変換
  im = plt.imshow(new_w, cmap='RdPu', # 色の指定
    interpolation='nearest', animated=True) # 動画の指定
  # ims には100枚の画像が一時的に保存
  ims.append([im]) # 画像をimsに保存
  old_w = new_w.copy() # 遷移した状態を次のステップで利用する準備
# 動画の作成  
ani = ani.ArtistAnimation(fig, # 画像の大きさなどを指定
       ims, # 画像を指定 
       interval=500, # 画像を切り替えるインターバル 
       blit=True, repeat_delay=1000) 
# 動画(mp4)としてdataフォルダに保存
ani.save('data//ca.mp4', writer='ffmpeg')     
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as ani
# nとfigsizeが大きくなればライフゲームの世界も大きくなる
fig = plt.figure(figsize=(5,5),dpi=150) # 画像の設定
n = 100         # 画像(配列)の大きさ
ims = []        # 描画した画像を一枚一枚保存する入れ物
np.random.seed(0) #  数値0を変更してみましょう
# 初期状態の作成： 要素を0.1の確率で1、0.9の確率で0に設定
old_w = np.random.binomial(1, 0.1, (n,n))
plt.tick_params(length=0,labelbottom=False,labelleft=False) # 余計な枠を非表示に設定

for step in np.arange(100): # 世代交代の回数
  new_w = np.zeros([n, n])  # 次の状態の入れ物を用意
  for i in np.arange(n): # 二次元配列の行の移動
    for j in np.arange(n): # 二次元配列の列の移動
      # 合計s に 上下左右、左上右上、左下右下の値を加算
      s = old_w[i - 1, j] # 上
      s += old_w[(i + 1) % n, j] # 下
      s += old_w[i, j - 1] # 左
      s += old_w[i, (j + 1) % n] # 右
      s += old_w[i - 1, j - 1] # 左上
      s += old_w[i - 1, (j + 1) % n] # 右上
      s += old_w[(i + 1) % n, j - 1] # 左下
      s += old_w[(i + 1) % n, (j + 1) % n] # 右下  
      # 合計sからold_w[i, j]の生物の次世代での生死を判定
      if old_w[i, j] == 0 and 3 == s: # ルールA
        new_w[i, j] = 1 # 適度な生物数のため、次は復活
      elif old_w[i, j] == 1 and 2 <= s <= 3: # ルールB
        new_w[i, j] = 1 # 適度な生物数のため、次も生存 
      elif old_w[i, j] == 1 and s <= 1 or 4 <= s: #ルールC
        new_w[i, j] = 0 # 過疎または過密化により次は死亡

  im = plt.imshow(new_w, cmap='RdPu', interpolation='none', animated=True) # 動画の指定
  ims.append([im]) # 画像をimsに保存
  old_w = new_w.copy() # 一つ前の状態として遷移した状態を次のステップで利用する準備
ani = ani.ArtistAnimation(fig, ims, interval=500, blit=True, repeat_delay=1000)
ani.save('lifegame.mp4', writer='ffmpeg')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as ani

fig = plt.figure(figsize=(10,10),dpi=300) # 画像の設定
m = 1000 # 画像の細かさ

real = np.linspace(-1.8,0.6,m) # 横軸に対応
imag = np.linspace(-1.2,1.2,m) # 縦軸に対応
C = np.zeros([m, m],dtype='complex')

# 複素数に変換
for i,real_d in enumerate(real):
    for j,imag_d in enumerate(imag):
        C[i,j] = np.complex(real_d, imag_d)
Z = np.zeros([m, m],dtype='complex') #描画する値を初期化
ims = [] # 描画した画像を一枚一枚保存する入れ物
# 配列にマンデルブロウ集合を描画するための計算結果を代入
for k in np.arange(50): #50枚描画する
    for i in np.arange(m):     # 横軸分繰り返し
        for j in np.arange(m): # 縦軸分繰り返し
            Z[i,j] = Z[i,j] ** 2 + C[i,j]
            if abs(Z[i,j]) >= 5: # 警告を回避するため
                Z[i,j] = k # ループ変数の値を最大値として代入
    im = plt.imshow(np.exp(-abs(Z.T)), cmap='magma',
         interpolation='nearest',animated=True) 
    ims.append([im])
# 動画ファイルを出力
ani = ani.ArtistAnimation(fig, ims, interval=500, blit=True, repeat_delay=1000)
ani.save('Mandelbrot_set.mp4', writer='ffmpeg')       
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as ani

fig = plt.figure(figsize=(10,10),dpi=300) # 画像の設定
m = 1000 # 画像の細かさ

real = np.linspace(0.4 , 0.415, m) 
imag = np.linspace(0.306,  0.312, m) 
# 複素数の行列 (このCがマンデルブロ集合となる)
C = np.zeros([m, m],dtype='complex')

# 複素数に変換
for i,real_d in enumerate(real):
    for j,imag_d in enumerate(imag):
        C[i,j] = np.complex(real_d, imag_d)
Z = np.zeros([m, m],dtype='complex') #描画する値を初期化
ims = [] # 描画した画像を一枚一枚保存する入れ物
# 配列にマンデルブロウ集合を描画するための計算結果を代入
for k in np.arange(60): #60枚描画する
    for i in np.arange(m):     # 横軸分繰り返し
        for j in np.arange(m): # 縦軸分繰り返し
            Z[i,j] = Z[i,j] ** 2 + C[i,j]
            if abs(Z[i,j]) >= 4:
                Z[i,j] = k
            elif abs(Z[i,j]) <= 0.001:
                Z[i,j] = -k
    im = plt.imshow(np.exp(-abs(Z.T)), cmap='magma',
         interpolation='nearest',animated=True) 
    ims.append([im])
# 動画ファイルを出力
ani = ani.ArtistAnimation(fig, ims, interval=500, blit=True, repeat_delay=1000)
ani.save('Mandelbrot_set.mp4', writer='ffmpeg')    
```
