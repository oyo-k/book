```python
#基礎課題13.1
import pandas as pd
df = pd.read_csv('data//seiseki_bj_ren.csv', index_col = 'ID')
for id in df.index:
    if df.at[id,'国語'] >= 80:
        df.at[id,'判定'] = 'A'
    elif df.at[id,'国語'] >= 60:
        df.at[id,'判定'] = 'B'
    else:
        df.at[id,'判定']= '不合格'
print(df)
df.to_csv('data//for_kadai_results.csv')
```

```python
# 発展課題13.3
import numpy as np
win_count = 0
for i in np.arange(0,3):
    my_hand = input('my_handの入力: rock, scissors, or, paper:')
    cpu_hand = input('cpu_handの入力: rock, scissors,  or, paper:')
    print('my_hand:', my_hand, '. cpu_hand:', cpu_hand)
    if my_hand == cpu_hand:
        print('Rock, paper, scissors')
    elif my_hand == 'rock' and cpu_hand == 'scissors':
        print('You win.')
        win_count = win_count + 1
    elif my_hand == 'scissors' and cpu_hand == 'paper':
        print('You win.')
        win_count = win_count + 1
    elif my_hand == 'paper' and cpu_hand == 'rock':
        print('You win.')
        win_count = win_count + 1
    else:
        print('You lose.')

print('あなたの勝利した回数', win_count)
```

```python
# 発展課題13.4
import pandas as pd
data = pd.read_csv('data//seiseki_5s_g.csv', index_col = 0)
print(data)
for sid in data.index:  
    data.loc[sid, '合計'] = data.loc[sid, '国語':'社会'].sum()
    if data.loc[sid, '合計'] >= 300: 
        data.loc[sid, '判定'] = '合格'
    else:
        data.loc[sid, '判定'] = '不合格'
print('成績評価後')
print(data)
data.to_csv('data//seiseki_5s_g_hantei.csv', index = True)
```

```python
from tkinter import Tk
import tkinter.simpledialog as sg, tkinter.messagebox as mx, numpy as np
root = Tk() # 描画を制御するおまじない
root.withdraw() # 余計なウィンドウを出さないため
# 入力画面を描画
name = sg.askstring('ELIZA:', 'あなたの名前を入力してください')
# 入力内容を使いメッセージを出力
mx.showinfo('ELIZA:', (name,'さん。よろしくお願いします。'))
for i in np.arange(3):
    # 会話内容の入力：問題文が指定するキーワードを利用
    data = sg.askstring('ELIZA:', '何かお話ししましょう')
    # コンピュータ (ELIZA) の行動をランダムに選択するため、乱数を生成
    selection = str(np.random.randint(0, 5)) 
    # コンピュータ (ELIZA) の会話パターンを if 文で作成
    if 'さようなら' == data: # 入力が'さようなら'ならば
          break # 繰り返し処理が繰り返し回数に関係なく終了
    elif '0' == selection: # 乱数 0 は セリフ 'さすがですね' に対応
          mx.showinfo('ELIZA:', name+'さん。さすがですね')
    elif '1' == selection: # 乱数 1 は セリフ '知らなかったです' に対応
          mx.showinfo('ELIZA:', name+'さん。知らなかったです')
    elif '2' == selection: # 乱数 2 は セリフ 'すごいですね' に対応
          mx.showinfo('ELIZA:', name+'さん。すごいですね')
    elif '3' == selection: # 乱数 3 は セリフ 'センスいいですね' に対応
          mx.showinfo('ELIZA:', name+'さん。センスいいですね')
    elif '4' == selection: # 乱数 4 は セリフ 'そうなんですか?' に対応
          mx.showinfo('ELIZA:', name+'さん。そうなんですか?')
    else:  # それ以外 セリフ '何か言いましたか？' に対応
          mx.showinfo('ELIZA:', name+'さん。何か言いましたか？')
# 入力された名前を使いメッセージを出力
mx.showinfo('ELIZA:', name+'さん。さようなら')
root.destroy() # 描画命令を終了
```

```python
# 発展課題13.6のプログラムそのまま
import numpy as np
for step in np.arange(3):
    print('Step', step, 'start!')
    my_hand = input('my_handの入力: rock, scissors, or, paper : ')
    # コンピュータの選択
    cpu_select =  np.random.randint(3)
    if cpu_select == 0:
        cpu_hand = 'paper'
    elif cpu_select == 1:
        cpu_hand = 'scissors'
    else:
        cpu_hand = 'rock'
    print('my_hand:', my_hand, '. cpu_hand:', cpu_hand)
    # 勝敗の判定
    if my_hand == cpu_hand:
        print('Rock, paper, scissors')
    elif my_hand == 'rock' and cpu_hand == 'scissors':
        print('You win.')
    elif my_hand == 'scissors' and cpu_hand == 'paper':
        print('You win.')
    elif my_hand == 'paper' and cpu_hand == 'rock':
        print('You win.')
    else:
        print('You lose.')
print('Game Over')
```
