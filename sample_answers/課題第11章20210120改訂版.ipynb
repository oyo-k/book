```python
# 基礎課題11.1
import numpy as np
import pandas as pd 
df_data = pd.DataFrame(
    {'a': np.arange(0,19,6),
     'b': np.arange(1,20,6),
     'c': np.arange(2,21,6),
     'd': np.arange(3,22,6),
     'e': np.arange(4,23,6),
     'f': np.arange(5,24,6)},
    index = [1,2,3,4])
display(df_data)
```

```python
# 基礎課題9.2
print('問1')
print(df_data.loc[3,'d'])
print(df_data.iloc[2,3]) #別解1
print(df_data.at[3,'d']) #別解2
print(df_data.iat[2,3]) #別解3

print('問2')
print(df_data.loc[1,:])
print(df_data.loc[1,]) # 別解1 (カラム名の場合コロンは省略化)
print(df_data.iloc[0,:]) # 別解2

print('問3')
print(df_data.loc[:,'d'])
print(df_data.iloc[:,3]) # 別解1

print('問4')
print(df_data.loc[2:5,'c':'e'])
```
