# 基礎課題20.2

```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt 

plt.rcParams['font.size'] = 15 # フォントの設定
plt.rcParams['lines.linewidth'] = 2
plt.rcParams['lines.markersize'] = 10

# 記述量を減らすため、共通するy軸のデータとインデックスのデータは変数に代入
yd = np.arange(20, 25) # y軸のデータ
id = np.arange(1, 6) # インデックスのデータ
# 星型マーカーのグラフのデータ
df1 = pd.DataFrame({'xdf':np.arange(10, 15), 'ydf':yd}, index = id)   
# 三角形のマーカーのグラフのデータ
df2 = pd.DataFrame({'xdf':np.arange(15, 20), 'ydf':yd}, index = id)   
# ひし形のマーカーのグラフのデータ
df3 = pd.DataFrame({'xdf':np.arange(20, 25), 'ydf':yd}, index = id)   

# 星型のマーカーのグラフを描画
plt.plot('xdf', 'ydf', color = 'b', marker = '*', linestyle = '', data = df1, label = 'line 1')
# 三角形のマーカーのグラフを描画
plt.plot('xdf', 'ydf', color = 'y', marker = '^', linestyle = '-', data = df2, label = 'line 2')
# ひし形のマーカーのグラフを描画
plt.plot('xdf', 'ydf', color = 'g', marker = 'D', linestyle = '--', data = df3, label = 'line 3')

# best を指定するため、凡例の位置は異なる場合があります。
plt.legend(loc = 'best')
plt.show() # 上記の描画したものを可視化
```

## 基礎課題19.2 の別解

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
plt.rcParams['font.size'] = 15 #フォントの設定
plt.rcParams['lines.linewidth'] = 2
plt.rcParams['lines.markersize'] = 10
# 一つ目のグラフのデータ
df1 = pd.DataFrame(
    {'xdf':np.arange(10,15), #  x軸のデータ
    'ydf':np.arange(20,25)}, #  y軸のデータ
    index=np.arange(1,6))   # インデックスをペアの名前として利用
# 二つ目のグラフのデータ
df2 = pd.DataFrame(
    {'xdf':np.arange(15,20), #  x軸のデータ
    'ydf':np.arange(20,25)}, #  y軸のデータ
    index=np.arange(1,6))   # インデックスをペアの名前として利用
# 三つ目のグラフのデータ
df3 = pd.DataFrame(
    {'xdf':np.arange(20,25), #  x軸のデータ
    'ydf':np.arange(20,25)}, #  y軸のデータ
    index=np.arange(1,6))   # インデックスをペアの名前として利用
# 一つ目のグラフを描画
plt.plot('xdf','ydf','b*', 
         data=df1,label='line 1')
# 二つ目のグラフを描画
plt.plot('xdf','ydf','y^-', 
         data=df2,label='line 2')
# 三つ目のグラフを描画
plt.plot('xdf','ydf','gd--',
         data=df3,label='line 3')
plt.legend(loc='best')
plt.show()
```

# 基礎課題20.4

```python
import pandas as pd
import matplotlib.pyplot as plt 
df = pd.read_csv('data//iris.csv',index_col="ID")
plt.rcParams['font.size'] = 12 #フォントの設定
plt.rcParams['figure.figsize'] = [4, 4]

plt.rcParams['lines.markersize'] = 10
plt.plot('Petal.Length','Petal.Width', 'g+', 
         data = df.loc[df['Species']=='setosa', :], label = 'setosa')
plt.plot('Petal.Length','Petal.Width', 'rx', 
         data = df.loc[df['Species']=='versicolor', :], label = 'versicolor')
plt.plot('Petal.Length','Petal.Width', 'b3', 
         data = df.loc[df['Species']=='virginica', :], label = 'virginica')
plt.legend(loc = 'best')
plt.xlabel('Petal.Length')
plt.ylabel('Petal.Width')
plt.savefig('data//iris//h19-2.pdf', bbox_inches = "tight", dpi = 300)
```
