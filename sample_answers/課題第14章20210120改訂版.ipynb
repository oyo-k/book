```python
# 基礎課題14.1
import numpy as np
from turtle import *
shape('turtle') # 亀を表示
speed(0)
penup() # ペン先を上げる
r = 0 # 最初に半径は 0 からスタートする
# 360度を1回転として5回転する
for th in np.arange(0, 1801, 6): 
  th = np.deg2rad(th)  # ラジアンに変換
  r = r + 0.5 # 半径を徐々に増加
  x = r * np.cos(th) # ペン先(x軸)の計算
  y = r * np.sin(th) # ペン先(y軸)の計算
  goto(x,y) # 移動しながら徐々に円を描く
  pendown() # ペン先を下げる
done()
```

```python
# 基礎課題14.2
from turtle import *
import numpy as np
import matplotlib.pylab as plt

p = 1 
q = 4 # ここを変更した
k = p / q  
cmap = plt.get_cmap('rainbow', (360 * q)) 
speed(0)
penup()
sized = 50
for i in np.arange(0, (360 * q) + 1, 1): 
    r = float(cmap(i)[0])
    g = float(cmap(i)[1])
    b = float(cmap(i)[2]) 
    pencolor(r, g, b) 
    th = np.deg2rad(i) 
    r = sized * np.cos(k * th) 
    x = r * np.cos(th) 
    y = r * np.sin(th)
    goto(x, y)
    pendown()
done()
```

```python
# 基礎課題14.3
from turtle import *
import numpy as np
speed(0) # 高速に描画
# 二つの自由パラメータの設定
n, d = 10, 15 
sized = 200 # 模様の大きさ
for th in np.arange(0, 361, 1):
    th = np.deg2rad(th)
    r = sized * np.sin(n * th * d)
    x = r * np.cos(th * d) 
    y = r * np.sin(th * d)
    goto(x, y)
done()
```
