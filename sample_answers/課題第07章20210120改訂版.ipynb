```python
# 基礎課題7.1
import numpy as np 
kadai = np.arange(1, 31) # 整数値1から30の要素の一次元配列
kadai = kadai.reshape(5,6) # reshape(行数,列数)
print('表7.1')
print(kadai) # 二次元配列の出力

print('出力例1')
print(kadai[2,2])

print('出力例2')
print(kadai[1:4,1:4])

print('出力例3')
print(kadai[0,:])

print('出力例4')
print(kadai[:,2])
```

```python
# 発展課題7.5
data1 = np.array([[1, 2, 3],[4, 5, 6]]) 
# 合計の場合
print('')
print(np.sum(data1,axis=(0))) 
print(np.sum(data1,axis=(1))) 
print(np.sum(data1)) 

# 平均
print('平均')
print(np.mean(data1,axis=(0))) 
print(np.mean(data1,axis=(1))) 
print(np.mean(data1)) 

#  中央値
print('中央値')
print(np.median(data1,axis=(0))) 
print(np.median(data1,axis=(1))) 
print(np.median(data1)) 

# 標準偏差
print('標準偏差')
print(np.std(data1,axis=(0))) 
print(np.std(data1,axis=(1))) 
print(np.std(data1)) 

# 分散
print('分散')
print(np.var(data1,axis=(0))) 
print(np.var(data1,axis=(1))) 
print(np.var(data1)) 

# 最大値
print('最大値')
print(np.max(data1,axis=(0)))
print(np.max(data1,axis=(1)))
print(np.max(data1))

# 最小値 
print('最小値 ')
print(np.min(data1,axis=(0))) 
print(np.min(data1,axis=(1))) 
print(np.min(data1)) 


# 最大値の要素番号
print('大値の要素番号')
print(np.argmax(data1,axis=(0)))
print(np.argmax(data1,axis=(1)))
print(np.argmax(data1))

# 最小値 の要素番号
print('最小値 の要素番号')
print(np.argmin(data1,axis=(0))) 
print(np.argmin(data1,axis=(1))) 
print(np.argmin(data1)) 

```

```python
# 発展課題7.6
# ゲーム開始のマルバツゲームの盤面を、2次元配列を用いて表現
d = np.array([['空', '空', '空'], ['空', '空', '空'], ['空', '空', '空']]) 
print(d) # '○' や '×' を入力（置く）前の盤面を出力

# 1 ターン目
d[1,1] = '○' # 盤面の中央に先手プレイヤーの'○'を入力に対応
print(d) # 1回目の先手プレイヤーの行動結果を出力
d[0,0] = '×' # 盤面の左上に後手プレイヤーの'×'を入力に対応
print(d) # 1回目の後手プレイヤーの行動結果を出力

# 2 ターン目
d[0, 2] = '○' # 右上
print(d) # 先手プレイヤーの行動結果を出力
d[2, 0] = '×' # 左下
print(d) # 後手プレイヤーの行動結果を出力 

# 3 ターン目
d[1, 0] = '○' # 中央左
print(d) # 先手プレイヤーの行動結果を出力
d[1, 2] = '×' # 中央右
print(d) # 後手プレイヤーの行動結果を出力

# 4 ターン目
d[0, 1] = '○' # 中央上
print(d) # 先手プレイヤーの行動結果を出力
d[2, 1] = '×' # 中央下
print(d) # 後手プレイヤーの行動結果を出力

# 5 ターン目
d[2, 2] = '○' # 右下
print(d) # 先手プレイヤーの行動結果を出力
```
