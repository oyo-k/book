# プログラム16.2

```python
import numpy as np
# 計算結果の出力に漢数字を使うために kanzi_dan を作成
# 繰り返し処理を簡単にするため、配列の要素番号0に'零'を用意
kanzi_dan = np.array(['零', '一', '二', '三', '四', '五', '六', '七', '八', '九'])

for dan in np.arange(1, 10): # 〇〇の段を設定するためのループ
    print(kanzi_dan[dan], 'の段の計算開始') # kanzi_dan[dan]は'一'から'九'に置き換わる
    
    for kazu in np.arange(1, 10): # 1×〇〇から9×〇〇の〇〇を置き換えるために利用
        print(kanzi_dan[dan], 'の段の', dan, '×', kazu, '=', dan *  kazu, 'です')
    print(kanzi_dan[dan], 'の段の計算終了')
    
print('九九の計算終了')
```

# プログラム16.2の簡略版

```python
import numpy as np
for dan in np.arange(1, 10):
    print(dan,'の段の計算開始')
    for kazu in np.arange(1, 10):
        print(dan, 'の段の', dan, '×', kazu, '=', dan *  kazu, 'です')
    print(dan,'の段の計算終了')
print('九九の計算終了')    
```

# プログラム16.2の別の書き方

```python
# パターン2 漢数字をアラビア数字に変換するライブラリ
import numpy as np
import unicodedata
kansuuzi = ['一', '二', '三', '四', '五', '六', '七', '八', '九']
for dan in kansuuzi:
    print(dan,'の段の計算開始')
    for kazu in np.arange(1, 10):
        suuzi_dan = unicodedata.numeric(dan) # 漢数字にアラビア数字に変換
        print(dan, 'の段の', suuzi_dan, '×', kazu, '=', suuzi_dan *  kazu, 'です')
    print(dan,'の段の計算終了')
print('九九の計算終了')
```

# プログラム16.2の別の書き方

```python
# 別の書き方
import numpy as np
import unicodedata

kansuuzi = ['一', '二', '三', '四', '五', '六', '七', '八', '九']
for dan in kansuuzi:
    print(dan,'の段の計算開始')
    for kazu in np.arange(1, 10):
        suuzi_dan = unicodedata.numeric(dan)
        print(dan, 'の段の', suuzi_dan, '×', kazu, '=', suuzi_dan *  kazu, 'です')
    print(dan,'の段の計算終了')
print('九九の計算終了')
```

# プログラム16.3

```python
d = np.array([[33, 44, 55, 66], 
              [199, 231, 333, 751]])
for i in np.arange(2): # iは行の移動に利用   
      for j in np.arange(4): # jは列の移動に利用    
            print('行(i)=', i, '列(j)=', j,
            'の要素は', d[i, j])
```

# プログラム16.4

```python
for i in np.arange(d.shape[0]): # d.shape[0]は行数
    for j in np.arange(d.shape[1]): # d.shape[1]は列数  
        print('行(i)=', i, '列(j)=', j, 'の要素は', d[i, j])
```

# プログラム16.5

```python
# 9 行 9 列の要素1の2次元配列を生成
d = np.ones((9, 9), dtype = 'int') 
# i を i 段としても利用
for i in np.arange(d.shape[0]): 
  # i 段にかける数値 j としても利用
  for j in np.arange(d.shape[1]):
    d[i, j] = (i + 1) * (j + 1)
print(d)
```
