# プログラム19.1

```python
import pandas as pd
import numpy as np
# データの読み込み
data = pd.read_excel('data//demographic.xlsx', index_col = 'ID')
display(data) # データ全体を出力
print(data.columns) # カラム名を出力
```

# プログラム19.3

```python
male = 0 # 男性の人数を代入する変数に初期値0を代入する
female = 0 # 女性の人数を代入する変数に初期値0を代入する

# ID を利用して男女別の人数カウントを開始するために回答者分繰り返し
for i in data.index: # i は 1 から 15000 まで変化       
    #print(i)
    if data.at[i, '性別'] == 1: # ID i の人が 1 （男性）なら
        male = male + 1 # male に 1 を加える(男性の人数が 1 増加)
    else: # それ以外。すなわち、ID i の人が 2 （女性）であれば
        female = female + 1 # femaleに 1 を加える(女性の人数が 1 増加)
# カウントの結果である男性人数と女性人数を表示
print('男性:', male, '女性:', female, '合計人数:', male + female)
```

# プログラム19.4

```python
male_age = 0 # 男性の年齢を加える変数 (全男性の合計年齢)
female_age = 0 # 女性の年齢を加える変数(全女性の合計年齢)

for i in data.index:
    if data.at[i,'性別'] == 1: # ID iの人が男性なら
        # 男性の年齢の合計にID iの年齢を加算
        male_age = male_age + data.at[i,'年齢']
    else: # ID i が女性なら、女性の年齢の合計にID i の年齢を加える
        female_age = female_age + data.at[i,'年齢']
# 年齢の総合計を人数で割り、男性と女性の平均年齢を計算
print('男性の平均年齢:', male_age / male,' 女性の平均年齢:', female_age / female)
```

# プログラム19.5

```python
ave_male_age = male_age / male # 男性の平均年齢
ave_female_age = female_age / female # 女性の平均年齢
var_male = 0 # 男性の年齢の分散を求める変数
var_female = 0 # 女性の年齢の分散を求める変数

for i in data.index: 
    if data.at[i, '性別'] == 1: # ID iの人が男性(1)なら
        # 男性の個々の年齢と平均年齢の差の二乗を加算
        var_male += (data.at[i, '年齢'] - ave_male_age) ** 2 
    else: # ID iの人が女性(2)なら、女性の個々の年齢と平均年齢の差の二乗を加算
        var_female += (data.at[i, '年齢'] - ave_female_age) ** 2
var_male /= male # 9行目で計算した差の二乗の合計 / 人数 = 男性の年齢の分散
var_female /= female # 11行目で計算した差の二乗の合計 / 人数 = 男性の年齢の分散
# 標準偏差は分散(var_maleまたはvar_female)の平方根
print('男性の分散:',var_male, '標準偏差', np.sqrt(var_male))
print('女性の分散:',var_female, '標準偏差', np.sqrt(var_female))
```

# プログラム19.6

```python
df_s = data.sort_values('年齢') # 昇順（小さい順）に並び替え
print('min:',df_s.iloc[0, 0]) # 年齢の0番目
print('max:',df_s.iloc[len(data) - 1, 0]) # 年齢の14999番目
```

# プログラム19.7

```python
n = len(df_s) # データの数
q = (n - 1) * 50 / 100 # 整数と小数部分の計算のために用意
i = int(q) # 整数に変換して整数部分を求める
dp = q - i # 実数から整数部分を除き小数部分を求める
# 「i番目の年齢 + 小数部分 (i+1番目の年齢 - i番目の年齢)」
q50 = df_s.iloc[i, 0] + dp * (df_s.iloc[i + 1, 0] - df_s.iloc[i, 0])
print('50%点の位置:',i , '小数:',dp , '中央値:', q50)
```

# プログラム19.22

```python
df = pd.DataFrame({'ren':np.arange(5, 8)})
display(df)
goukei = 0 # 合計を入れる変数
var = 0 # 分散を入れる変数
for k in df.index: # 平均値を求めるために合計を算出
    goukei = goukei + df.at[k, 'ren']
ave = goukei / df.size # 平均値の計算
for j in df.index: # 分散の計算のために繰り返し
    var = var + (df.at[j, 'ren'] - ave) ** 2
var = var / df.size # 分散を計算
print('平均:', ave, '分散:', var) 
```

# プログラム19.23

```python
df_ren = pd.DataFrame({'ren':np.arange(10, 20)})
df_rens = df_ren.sort_values('ren') # 並び替え
print('min:', df_rens.iloc[0, 0], '便利な関数:',df_ren['ren'].min()) # 最小値
print('max:', df_rens.iloc[len(df_rens) - 1, 0], '便利な関数:', df_ren['ren'].max()) # 最大値
q = (len(df_rens) - 1) * 25 / 100  # 実数値を求める
i = int(q) # 整数部分
dp = q - i # 小数部分
q25 = df_rens.iloc[i, 0] + dp * (df_rens.iloc[i+1, 0] - df_rens.iloc[i, 0])
print('25%点の位置:', i , '小数:', dp, '25%点:',q25)
print('50%点は', df_ren['ren'].quantile(0.50), '75%点は', df_ren['ren'].quantile(0.75))
```
